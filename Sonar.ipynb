```python
import scipy
import scipy as stats
from scipy.stats import zscore
from scipy.stats import percentileofscore
z=np.abs(zscore(main_df))
new_df=main_df[(z<3).all(axis=1)]
```

```python
print("original shape=",main_df.shape,"\nNew shape=",new_df.shape)
```

```python
new_df.skew()
```

```python
x=new_df.iloc[:,:-1]
y=new_df.iloc[:,-1]
```

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method=("yeo-johnson"))
```

# Model Fitting and Predicting

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

To evaluate our model performance, we can use the make_scorere and accuracy_score function from sklearn metrics.

```python
from sklearn.metrics import make_scorer, accuracy_score
```

We can also use a GridSearch cross validation to find the optimal parameters for the model we choose to work with and use to predict on our testing set.

```python
from sklearn.model_selection import GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=42)
```

```python
svc_clf = SVC() 
rf_clf = RandomForestClassifier()
logreg_clf = LogisticRegression()
knn_clf = KNeighborsClassifier()
gnb_clf = GaussianNB()
dt_clf = DecisionTreeClassifier()
list1=[svc_clf,rf_clf,logreg_clf,knn_clf,gnb_clf,dt_clf ]
for i in list1:
    i.fit(x,y)
    pred = i.predict(x_test)
    acc = accuracy_score(y_test, pred)

    print(acc)
```

```python
rf_clf = RandomForestClassifier()

parameters = {"n_estimators": [4, 5, 6, 7, 8, 9, 10, 15], 
              "criterion": ["gini", "entropy"],
              "max_features": ["auto", "sqrt", "log2"], 
              "max_depth": [2, 3, 5, 10], 
              "min_samples_split": [2, 3, 5, 10],
              "min_samples_leaf": [1, 5, 8, 10]
             }

grid_cv = GridSearchCV(rf_clf, parameters, scoring = make_scorer(accuracy_score))
grid_cv = grid_cv.fit(x_train, y_train)

print("Our optimized Random Forest model is:")
grid_cv.best_estimator_
```

```python
rf_clf = grid_cv.best_estimator_

rf_clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import plot_confusion_matrix

disp = plot_confusion_matrix(rf_clf, x_train, y_train,
                             display_labels=['ROCK','MINE'],
                             cmap=plt.cm.Reds,
                             normalize=None)
disp.ax_.set_title('Confusion matrix')

print('Train results: confusion matrix')
print(disp.confusion_matrix)
```
